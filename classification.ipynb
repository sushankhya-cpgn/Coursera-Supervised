```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])
x_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
pos = y_train == 1
neg = y_train == 0
fig,ax = plt.subplots(1,2,figsize = (8,3))
ax[0].scatter(x_train[pos],y_train[pos],marker='x',s=80,c='r',label='y=1')
ax[0].scatter(x_train[neg],y_train[neg],marker='o',s=80,c='b',label='y=0')
ax[0].set_ylabel('y',fontsize=12)
ax[0].set_xlabel('x',fontsize=12)
ax[0].set_title("One variable plot")
ax[0].legend();

# Plot two variables
pos = y_train2 == 1
neg = y_train2 == 0
print(pos)
print(neg)
ax[1].axis([0,4,0,4])
ax[1].scatter(x_train2[pos,0],x_train2[pos,1],marker='x',s=80,c='r',label='y=1')
ax[1].scatter(x_train2[neg,0],x_train2[neg,1],marker='o',s=80,c='b',label='y=0')
ax[1].set_ylabel('x2',fontsize=12)
ax[1].set_xlabel('x1',fontsize=12)
ax[1].set_title("Two variable plot")
ax[1].legend();
plt.show();
```
