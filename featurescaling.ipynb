```python
import numpy as np
import matplotlib.pyplot as plt
x_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460,232,178])
b_init = 785.1811367994083
w_init = np.array([ 0.39133535, 18.75376741, -53.36032453, -26.42131618])
```

```python
def zscore_normalize_feature(X):
    mu = np.mean(X,axis = 0)
    sigma = np.std(X,axis=0)
    X_norm = (X-mu)/sigma
    return X_norm,mu,sigma
```

```python
X_norm,X_mean,X_sigma = zscore_normalize_feature(x_train)
print(f"X_mu = {X_mean}, \nX_sigma = {X_sigma} \n \n X_norm {X_norm}")
fig,ax = plt.subplots(1,2,figsize = (12,6))
ax[0].scatter(x_train[:,1],x_train[:,3],s=50)
ax[0].set_title("Before Normalization")
ax[1].scatter(X_norm[:,1],X_norm[:,3],s=50)
ax[1].set_title("After Normalization")
plt.tight_layout(rect=[0, 0.02, 1, 0.9])
fig.suptitle("Distribution of features before, after normalization")
plt.show()

```
